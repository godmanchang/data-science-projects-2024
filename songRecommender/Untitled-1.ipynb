```python
# Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
# Load Iris dataset
data = datasets.load_iris()
df = pd.DataFrame(data['data'])
df.columns = data['feature_names']
df['target'] = data['target']
```
